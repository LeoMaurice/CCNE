```python
import bibtexparser
import pandas as pd
from bibtexparser.bparser import BibTexParser
from bibtexparser.customization import convert_to_unicode

def bibtex_to_csv(bibfile_path):
    with open(bibfile_path) as bibtex_file:
        parser = BibTexParser(common_strings=True)
        parser.customization = convert_to_unicode
        bib_database = bibtexparser.load(bibtex_file, parser=parser)
        
    # Convert to DataFrame
    df = pd.DataFrame(bib_database.entries)
    return df

# Example usage
bibfile_path = 'anthology+abstracts.bib'
csv_file_path = 'acl_proceedings.csv'
df = bibtex_to_csv(bibfile_path)
df
```

```python
df['title'] = df['title'].replace({
    ';': ' ',               # Replace semicolons with space
    '"': " "              # Replace double quotes with two single quotes
}, regex=True)
```

```python
df['abstract'] = df['abstract'].replace({
    ';': ' ',               # Replace semicolons with space
    '"': " "              # Replace double quotes with two single quotes
}, regex=True)
```

```python
df['author'] = df['author'].replace({
    ' ': '',
    'and': '',
    "\n":"***"
}, regex=True)
```

```python
df['year'] = df['year'].astype(int)
```

```python
df_with_abstract = df[["ID","author","year","title","abstract"]].dropna()
df_with_abstract.sample(10)
```

```python
# Save to CSV
df[["ID","author","year","title","abstract"]].dropna().to_csv(csv_file_path, index=False, quoting=2,sep=";")
```
